```python
import copy
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from scipy.optimize import curve_fit
```

```python
# fxn for plotting
def get_ax(figsize=(6,4), fsize=15):
    fig, ax = plt.subplots(figsize=figsize)
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    return ax, fsize
```

```python
# hot
run1 = pd.read_csv("run1.csv")
# cold
run2 = pd.read_csv("run2.csv")
```

```python
# R_p = 10k ohms, R_s = 33 ohms
# v = ir
# i = v/r
run1["Ip"] = run1["Vp"] / 10**4
run2["Ip"] = run2["Vp"] / 10**4
```

```python
ax, fsize = get_ax()

plt.plot(np.sqrt(run1["V - Vs"]), run1["Ip"]*10**6, c="0", markersize=fsize-5, marker=".")

plt.xticks(size=fsize-2)
plt.yticks(size=fsize-2)
plt.xlabel("$\sqrt{V-V_s}$ ($\sqrt{V}$)", size=fsize)
plt.ylabel("$I_p\; (\mu\;A)$", size=fsize)

# plt.legend(fontsize=fsize, frameon=False)

plt.tight_layout()
plt.savefig("Plots/hot.pdf")
```

```python
ax, fsize = get_ax()

plt.plot(np.sqrt(run2["V - Vs"]), run2["Ip"]*10**6, c="0", markersize=fsize-5, marker=".")

plt.xticks(size=fsize-2)
plt.yticks(size=fsize-2)
plt.xlabel("$\sqrt{V-V_s}$ ($\sqrt{V}$)", size=fsize)
plt.ylabel("$I_p^*\; (\mu\;A)$", size=fsize)

# plt.legend(fontsize=fsize, frameon=False)

plt.tight_layout()
plt.savefig("Plots/cold.pdf")
```

```python
ax, fsize = get_ax()

plt.plot(np.sqrt(run1["V - Vs"]), run1["Ip"]*10**6, c="0", markersize=fsize-5, marker=".", label="$I_p$")
plt.plot(np.sqrt(run2["V - Vs"]), run2["Ip"]*10**6, c="r", markersize=fsize-5, marker=".", label="$I_p^*$")

plt.xticks(size=fsize-2)
plt.yticks(size=fsize-2)
plt.xlabel("$\sqrt{V-V_s}$ ($\sqrt{V}$)", size=fsize)
plt.ylabel("$\mu\;A$", size=fsize)

plt.legend(fontsize=fsize, frameon=False)

plt.tight_layout()
plt.savefig("Plots/both.pdf")
```

```python
num = run1["Vp"] * run2["Vs"]
den = run1["Vs"] * run2["Vp"]
T = num / den
```

```python
ax, fsize = get_ax()

plt.plot(run1["V - Vs"], T*100, c="0", markersize=fsize-5, marker=".")

x_max = run1["V - Vs"][np.argmax(T)]
ax.vlines(x_max, 0, np.max(T)*100, color="r")
x_max_r = np.round(x_max, 3)

plt.text(x_max, 0, f"$V-V_s={x_max_r},\;T=93\%$", color="r")

plt.xticks(size=fsize-2)
plt.yticks(size=fsize-2)
plt.xlabel("$V-V_s$ (V)", size=fsize)
plt.ylabel("$T\;(\%)$", size=fsize)

# plt.legend(fontsize=fsize, frameon=False)

plt.tight_layout()
plt.savefig("Plots/T.pdf")
```

```python
# hot
run3 = pd.read_csv("run3.csv")
run3["Vs"] *= 10**6
# cold
run4 = pd.read_csv("run4.csv")
run4["Vs"] *= 10**6
```

```python
run4.head()
```

```python
def linear(x, m, b):
    return m*x + b
```

```python
ax, fsize = get_ax()

params, param_covs = curve_fit(linear, run3["V - Vs"], run3["Vs"])
print(params)
m, b = params
fit_x = np.arange(np.min(run3["V - Vs"]), np.max(run3["V - Vs"]), 10**(-2))
fit_y = [linear(x, m, b) for x in fit_x]

plt.plot(run3["V - Vs"], run3["Vs"], c="0", markersize=fsize-5, marker=".", label="data")

m_r = np.round(m, 3)
b_r = np.round(b, 3)
plt.plot(fit_x, fit_y, c="0", linestyle=":", label=f"$y=mx+b$\n$m={m_r}$\n$b={b_r}$")

x_zero = fit_x[np.argmin(np.abs(fit_y))]
plt.vlines(x_zero, np.min(run3["Vs"]), np.max(run3["Vs"]), color="r", linestyle="-.")
plt.hlines(0, np.min(run3["V - Vs"]), np.max(run3["V - Vs"]), linestyle="-.", color="r")

x_zero_r = np.round(x_zero, 3)
plt.text(x_zero+0.01, -1.9, f"$V-V_s=${x_zero_r}", color="r")
plt.text(-3.1, 0.1, "$V_s=0$", color="r")


plt.xticks(size=fsize-2)
plt.yticks(size=fsize-2)
plt.xlabel("$V-V_s$ (V)", size=fsize)
plt.ylabel("$V_s$ ($\mu V$)", size=fsize)

plt.legend(fontsize=fsize, frameon=False)

plt.tight_layout()
plt.savefig("Plots/hot_offset.pdf")
```

```python
ax, fsize = get_ax()

params, param_covs = curve_fit(linear, run4["V - Vs"], run4["Vs"])
print(params)
m, b = params
fit_x = np.arange(np.min(run4["V - Vs"]), np.max(run4["V - Vs"]), 10**(-2))
fit_y = [linear(x, m, b) for x in fit_x]

plt.plot(run4["V - Vs"], run4["Vs"], c="0", markersize=fsize-5, marker=".", label="data")

m_r = np.round(m, 3)
b_r = np.round(b, 3)
plt.plot(fit_x, fit_y, c="0", linestyle=":", label=f"$y=mx+b$\n$m={m_r}$\n$b={b_r}$")

x_zero = fit_x[np.argmin(np.abs(fit_y))]
plt.vlines(x_zero, np.min(run4["Vs"]), np.max(run4["Vs"]), color="r", linestyle="-.")
plt.hlines(0, np.min(run4["V - Vs"]), np.max(run4["V - Vs"]), linestyle="-.", color="r")

x_zero_r = np.round(x_zero, 3)
plt.text(x_zero+0.01, -1.7, f"$V-V_s=${x_zero_r}", color="r")
plt.text(-3, 0.1, "$V_s=0$", color="r")


plt.xticks(size=fsize-2)
plt.yticks(size=fsize-2)
plt.xlabel("$V-V_s$ (V)", size=fsize)
plt.ylabel("$V_s$ ($\mu V$)", size=fsize)

plt.legend(fontsize=fsize, frameon=False)

plt.tight_layout()
plt.savefig("Plots/cold_offset.pdf")
```
